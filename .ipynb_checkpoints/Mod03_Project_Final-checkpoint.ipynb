# Hypothesis 1: East Score vs West Score

**Null Hypothesis**
-  $H_{0}: \mu_{east} = \mu_{west}$

**Alternative Hypothesis**

- $H_{1}: \mu_{east} > \mu_{west}$

## Paired t-test

Since each row is a game and each game has 2 variables of interest (East team score and West team score), we want to examine the difference in scores. We will conduct the paired t-test to compare the means of the East Conference team scores and West Conference team scores to see if there is a significant difference between the two scores (vs. there being no difference, East team score - West team score = 0).

### Hypothesis Test Results

```python
tests.conduct_pairedttest(east_west_games,
                          'east_score',
                          'west_score')
```

### Hypothesis Test Visualization

```python
v.make_box_plot(east_west_games[['east_score', 'west_score']],
                east_west_games[['east_minus_west']],
                'distribution_of_points')
```

# Hypothesis 2: Which Team has Better Home Court Advantage

**Null Hypothesis**
- $H_{0}: \mu_{team1} = \mu_{team2} = \cdots = \mu_{n}$

**Alternative Hypothesis**
- $H_{1}$: at least one team has home court advantage

## Shape dataframes for test

```python
#west_teams
west_home_game_data = dp.make_home_df(games_data, 'West', 7)
west_home_game_data
```

```python
#east_teams
east_home_game_data = dp.make_home_df(games_data, 'East', 7)
east_home_game_data
```

## ANOVA Test

We want to compare a group of teams to see which team(s) has a significantly different home court (dis)advantage if one exists. We define home court advantage as the mean difference in score when a team plays a home game. We'll first conduct an ANOVA test to see if there exists a significant mean difference among teams. If there exists a significant mean, we'll conduct a ttest among all pairs of teams to compare their mean differences and identify the team(s) that led to our significant results in the ANOVA test.

### West Conference Teams

```python
tests.conduct_anova(west_home_game_data,
                    "home_team_full_name",
                    "home_score_diff")
```

```python
v.make_ordered_boxplot(west_home_game_data,
                       'home_team_full_name',
                       'home_score_diff',
                       'west_home_games')
```

### East Conference Teams

```python
tests.conduct_anova(east_home_game_data,
                    'home_team_full_name',
                    'home_score_diff')
```
